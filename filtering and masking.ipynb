```python
%sql
use catalog `main`;
use schema `dbdemos_aibi_cme_marketing_campaign`;
```

```python
%sql
CREATE FUNCTION country_filter(country STRING)
RETURN IF(IS_ACCOUNT_GROUP_MEMBER('admin'), true, country='NL');
```

```python
%sql
ALTER TABLE main.dbdemos_aibi_cme_marketing_campaign.prospects SET ROW FILTER country_filter ON (country);
```

```python
%sql
select distinct country from main.dbdemos_aibi_cme_marketing_campaign.prospects
```
